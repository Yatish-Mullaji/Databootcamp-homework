# Importing Modules and Dependencies 

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
#print(mouse_drug_df.head())

clinical_df = pd.read_csv(clinical_trial_data_to_load) 
#print(clinical_df.head())


# Combine the data into a single dataset
merged_mouse_drug_clinical_df = clinical_df.merge(mouse_drug_df,on="Mouse ID")


# Display the data table for preview
merged_mouse_drug_clinical_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_vol = merged_mouse_drug_clinical_df.groupby(["Drug","Timepoint"])

# Convert to DataFrame
mean_tumor_vol_df = pd.DataFrame(mean_tumor_vol["Tumor Volume (mm3)"].mean()).reset_index()

# Preview DataFrame
mean_tumor_vol_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
std_error_tumor_vol = mean_tumor_vol["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
std_error_tumor_vol_df = pd.DataFrame(std_error_tumor_vol).reset_index()

# Preview DataFrame
std_error_tumor_vol_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
munged_data_df = mean_tumor_vol_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
munged_data_df
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()
x_axis = munged_data_df.index

standard_errors_capomulin = std_error_tumor_vol_df.loc[std_error_tumor_vol_df["Drug"]=="Capomulin","Tumor Volume (mm3)"]
standard_errors_capomulin = std_error_tumor_vol_df.loc[std_error_tumor_vol_df["Drug"]=="Infubinol","Tumor Volume (mm3)"]
standard_errors_capomulin = std_error_tumor_vol_df.loc[std_error_tumor_vol_df["Drug"]=="Ketapril","Tumor Volume (mm3)"]
standard_errors_capomulin = std_error_tumor_vol_df.loc[std_error_tumor_vol_df["Drug"]=="Placebo","Tumor Volume (mm3)"]

ax.errorbar(x_axis, munged_data_df["Capomulin"],standard_errors_capomulin, fmt="o",linestyle='dotted',label="capomulin",color='r')
ax.errorbar(x_axis, munged_data_df["Infubinol"],standard_errors_capomulin, fmt="^",linestyle='dotted',label="Infubinol",color='b')
ax.errorbar(x_axis, munged_data_df["Ketapril"],standard_errors_capomulin, fmt="s",linestyle='dotted',label="Ketapril",color='g')
ax.errorbar(x_axis, munged_data_df["Placebo"],standard_errors_capomulin, fmt="D",linestyle='dotted',label="Placebo",color='black')

ax.set_xlim(std_error_tumor_vol_df["Timepoint"].min() - 3, std_error_tumor_vol_df["Timepoint"].max() + 3)

ax.set_xlabel("Time (Days)")
ax.set_ylabel("Tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")
plt.legend(loc="best")
y=np.linspace(35,75,9)
xmin=-5
xmax=70
plt.hlines(y,xmin,xmax,colors='k', linestyles='solid',alpha=0.75,linewidth=0.25)

# Save the Figure
plt.savefig("../Images/Tumor_Response_to_treatment.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count =merged_mouse_drug_clinical_df.groupby(["Drug","Timepoint"])

# Convert to DataFrame
mice_count_df = pd.DataFrame(mice_count["Mouse ID"].count()).reset_index()
mice_count_df = mice_count_df.rename(columns = {"Mouse ID":"Mouse Count"})

# Preview DataFrame
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
munged_datamice_count_df = mice_count_df.pivot(index='Timepoint', columns='Drug', values="Mouse Count")

# Preview the Data Frame
munged_datamice_count_df
```

```python
# Generate the Plot (Accounting for percentages)
#fig, ax3 = plt.subplots()
x_axis = munged_datamice_count_df.index

plt.plot(x_axis, (munged_datamice_count_df["Capomulin"]/25)*100, marker="o",linestyle='dotted',label="capomulin",color='r')
plt.plot(x_axis, (munged_datamice_count_df["Infubinol"]/25)*100, marker="^",linestyle='dotted',label="Infubinol",color='b')
plt.plot(x_axis, (munged_datamice_count_df["Ketapril"]/25)*100, marker="s",linestyle='dotted',label="Ketapril",color='g')
plt.plot(x_axis, (munged_datamice_count_df["Placebo"]/25)*100, marker="D",linestyle='dotted',label="Placebo",color='black')

#ax2.set_xlim(mice_count_df["Timepoint"].min() - 3, mice_count_df["Timepoint"].max() + 3)

plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.title("Survival During Treatment")
plt.legend(loc="best")
plt.grid()

# Save the Figure
plt.savefig("../Images/Survival During Treatment.png")
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_per_change = ((munged_data_df.loc[45,:]-munged_data_df.loc[0,:])/(munged_data_df.loc[0,:]))*100


# Display the data to confirm
tumor_per_change
```

```python
# Store all Relevant Percent Changes into a Tuple
tumor_per_change_tuple = tuple(zip(tumor_per_change.index,tumor_per_change))
tumor_per_change_tuple

# Splice the data between passing and failing drugs
tumor_per_change_tuple_df = pd.DataFrame(list(tumor_per_change_tuple),columns=["Drug","Percentage Change"])
tumor_per_change_tuple_df
tumor_per_change_tuple_df = tumor_per_change_tuple_df.drop(tumor_per_change_tuple_df.index[[1,4,6,7,8,9]]).reset_index()
tumor_per_change_tuple_df

# Orient widths. Add labels, tick marks, etc. 
bars=plt.bar(tumor_per_change_tuple_df["Drug"],tumor_per_change_tuple_df["Percentage Change"],align='edge',width=-1,
                                        color=(tumor_per_change_tuple_df["Percentage Change"]>0).map({True:'r',False:'g'}))


                                           
plt.grid()
plt.ylabel("% Tumor Volume Change")
plt.title("Tumor Change Over 45 Day Treatment")

# Use functions to label the percentages of changes
def per_label(label):
    
    for i,bar in enumerate(bars):
        plt.text(bar.get_x() + bar.get_width()/2., 0.1*bar.get_height(),
                   str(round(tumor_per_change_tuple_df["Percentage Change"][i],2))+"%",ha='center', va='top',color='w',alpha=25)

# Call functions to implement the function calls
per_label(bars)

# Save the Figure
plt.savefig("../Images/Tumor Change Over 45 Day Treatment.png")

# Show the Figure
fig.show()
```

# Obsevations:



The current study is conducted for 10x experimental cancer drugs.  Each drug was tested on the set of 25x rats for 45 days with the initial cancer tumour vol of 45mm^3 and observations were done every 5 days.
For the close observation of the experiment we have conducted detailed study on 4 drugs (Capomulin, Infubinol, Ketapril, Placebo). From the detailed observation following can be concluded:

	During the 45 days periods of time it was observed the tumour was keep on growing linearly from its initial condition for    all the drugs but Capomulin. The tumour was observed to be shrunk for the rats which were given Capomulin. 

	It was observed that Capomulin was the most effective drug when comes to metastatic condition of the tumour followed by  Infubinol.  The least effective drugs were Ketapril & Placebo.

	In terms of survival rate Infubinol proven to be the least effective with survival rate of less than 40% at the end of 45 days treatment. The most effective drug was Capomulin with survival rate of more than 85%.

	Overall, it can be concluded that only drug Capomulin  is proven to be effective based on this study.
