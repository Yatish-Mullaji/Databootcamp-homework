## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,70]
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
unique_names = unique_df["SN"]
total_gamers = unique_df["SN"].count()
unique_age = unique_df["Age"]
age_demographic_df = pd.DataFrame({"Total Count":unique_names,"Age":unique_age})
age_demographic_df.head()

age_demographic_df["Age Ranges"] = pd.cut(age_demographic_df["Age"], bins, labels=group_names)
age_demographic_df = age_demographic_df.groupby("Age Ranges").count()

age_demographic_df
age_demographic_table = pd.DataFrame(age_demographic_df["Total Count"])
age_demographic_table["Percentage of Players"] = (age_demographic_table["Total Count"]/total_gamers)*100
age_demographic_table["Percentage of Players"] = age_demographic_table["Percentage of Players"].map("{:,.2f}%".format)
del age_demographic_table.index.name
age_demographic_table
```

## Purchasing Analysis (Age)

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
total_client = purchase_data["SN"]
overall_gamers = purchase_data["SN"].count()
overall_age = purchase_data["Age"]
total_purchase_value = purchase_data["Price"]
purchase_analysis_df = pd.DataFrame({"Purchase Count":total_client,"Age":overall_age})
purchase_analysis_df.head()

purchase_analysis_df["Age Ranges"] = pd.cut(purchase_analysis_df["Age"], bins, labels=group_names)
purchase_analysis_df = purchase_analysis_df.groupby("Age Ranges").count()

purchase_analysis_df


purchase_analysis_table1 = pd.DataFrame(purchase_analysis_df["Purchase Count"])



purchase_analysis_table1

purchase_analysis_df2 = pd.DataFrame({"Age":overall_age,"Total Purchase Value":total_purchase_value})
purchase_analysis_df2.head()
purchase_analysis_df2["Age Ranges"] = pd.cut(purchase_analysis_df2["Age"], bins, labels=group_names)
purchase_analysis_df2 = purchase_analysis_df2.groupby("Age Ranges").sum()

purchase_analysis_df2

purchase_analysis_table2 = pd.DataFrame(purchase_analysis_df2["Total Purchase Value"])

purchase_analysis_table2

purchase_analysis_table = pd.merge(purchase_analysis_table1, purchase_analysis_table2, on="Age Ranges")
purchase_analysis_table["Average Purchase Price"] = purchase_analysis_table["Total Purchase Value"]/purchase_analysis_table1["Purchase Count"]
purchase_analysis_table["Avg Total Purchase per Person"] = purchase_analysis_table["Total Purchase Value"]/age_demographic_table["Total Count"]
purchase_analysis_table["Total Purchase Value"]=purchase_analysis_table["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_table["Average Purchase Price"]=purchase_analysis_table["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis_table["Avg Total Purchase per Person"]=purchase_analysis_table["Avg Total Purchase per Person"].map("${:,.2f}".format)
purchase_analysis_table
purchase_analysis_table = purchase_analysis_table.reindex(columns=["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"])
purchase_analysis_table
```

## Top Spenders
